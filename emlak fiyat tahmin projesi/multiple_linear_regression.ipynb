# MULTIPLE LINEAR REGRESSION

Multiple Linear Regression'da birden fazla bağımsız(independent) değişkene karşılık bir bağımlı(dependent) dğeişken bulunur.

Linear Regression veriler arasında var olan korelasyonu(ilişkiyi) kullanarak yeni gelecek verileri tahmin etme modelidir. Burada makine öğrenimi bize veriler arasındaki bu ilişkiyi belirlememize yardımcı olur ve bu sayede yeni verileri tahmin edebiliriz. 

```python

import pandas as pd
import matplotlib.pyplot as plt
# sklearn library
from sklearn import linear_model

# veri setimizi import ediyoruz, ayraç olarak noktalı virgül olduğu için bunu belirtiyoruz:
df = pd.read_csv("emlakveri.csv",sep = ";")

```

```python
# linear regression modeli tanımlıyoruz:

reg = linear_model.LinearRegression()
reg.fit(df[['BinaYasi', 'Metrekare','Kat']], df['Fiyat'])

# Prediction yapalım..
```

```python

reg.predict([[0,80,0]]) 
```

```python
reg.coef_
```

```python
import statsmodels.api as sm   # Modelin hata tespitinin yapıldığı kısım

X = sm.add_constant(df[['BinaYasi', 'Metrekare','Kat']])
model = sm.OLS(df['Fiyat'], X).fit()
model.summary()
```

```python
# Test ve Train olarak verinin ayrılması ve buna göre eğitilmesi
from sklearn.model_selection import train_test_split
#Humidty ---> Bağımlı değişken
#Diğer kolonlar ---> Bağımsız değişken
x_train, x_test,y_train,y_test = train_test_split(df[['BinaYasi', 'Metrekare','Kat']],df[['Fiyat']],test_size=0.33, random_state=0)
#sonveri.iloc[:,:-1]---> Bağımsız değişken son kolon hariç
#sonveri.iloc[:,-1:]---> Bağımlı değişken son kolon 

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)

y_pred = regressor.predict(x_test)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
import tkinter as tk
from tkinter import messagebox


class LinearRegressionGUI:
    def __init__(self, master):
        self.master = master
        master.title("Emlak Fiyat Tahmin Programı")
        master.geometry("600x600")

        # Başlık
        self.label_title = tk.Label(master, text="Emlak Fiyatı Tahmin Programı", font=("Helvetica", 18),fg="red")
        self.label_title.pack(pady=20)

        # Bina Yaşı etiketi ve giriş kutusu
        self.label_bina_yasi = tk.Label(master, text="Bina Yaşı:", font=("Helvetica", 14),fg="blue",bg="#F0F8FF")
        self.label_bina_yasi.pack(pady=10)
        self.entry_bina_yasi = tk.Entry(master, font=("Helvetica", 14),fg="blue",bg="#F0F8FF")
        self.entry_bina_yasi.pack(pady=10)

        # Metrekare etiketi ve giriş kutusu
        self.label_metrekare = tk.Label(master, text="Metrekare:", font=("Helvetica", 14),fg="blue",bg="#F0F8FF")
        self.label_metrekare.pack(pady=10)
        self.entry_metrekare = tk.Entry(master, font=("Helvetica", 14),fg="blue",bg="#F0F8FF")
        self.entry_metrekare.pack(pady=10)
        """
        #Metrekare combobox tarzı
        self.label_metrekare = tk.Label(master, text="Metrekare:", font=("Helvetica", 14), fg="blue", bg="#F0F8FF")
        self.label_metrekare.pack(pady=10)
        
        self.entry_metrekare = ttk.Combobox(master, font=("Helvetica", 14), state="readonly")
        self.entry_metrekare['values'] = ['1', '2', '3', '4', '5']
        self.entry_metrekare.pack(pady=10)
        """

        # Kat etiketi ve giriş kutusu
        self.label_kat = tk.Label(master, text="Bulunduğu Kat:", font=("Helvetica", 14), fg="blue", bg="#F0F8FF")
        self.label_kat.pack(pady=10)
        
        self.entry_kat = ttk.Combobox(master, font=("Helvetica", 14), state="readonly")
        
        self.entry_kat['values'] = ['0','1', '2', '3', '4', '5']
        self.entry_kat.pack(pady=10)
        style = ttk.Style()

        # Tahmin Et butonu
        self.button = tk.Button(master, text="Tahmin Et", command=self.predict, font=("Helvetica", 14),fg="blue",bg="#F0F8FF")
        self.button.pack(pady=20)

        # Fiyat sonucu etiketi
        self.label_result = tk.Label(master, text="Fiyat:", font=("Helvetica", 14),fg="blue",bg="#F0F8FF")
        self.label_result.pack(pady=10)

    def predict(self):
        # veri setini yükle
        df = pd.read_csv("emlakveri.csv",sep = ";")

        # linear regression modeli tanımla
        reg = linear_model.LinearRegression()
        reg.fit(df[['BinaYasi', 'Metrekare','Kat']], df['Fiyat'])

        # Prediction yap
        bina_yasi = float(self.entry_bina_yasi.get())
        metrekare = float(self.entry_metrekare.get())
        kat = float(self.entry_kat.get())
        fiyat = reg.predict([[bina_yasi, metrekare, kat]])

        # sonucu ekrana yaz
        self.label_result.config(text="Tahmini Fiyat: " + str(round(fiyat[0], 2)),fg="green")


root = tk.Tk()
linear_regression_gui = LinearRegressionGUI(root)
root.mainloop()
```
